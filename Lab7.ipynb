```python
import pandas as pd
import numpy as np
```

```python
f = pd.read_csv("censusCrimeClean.csv")
f
```

```python
f = f.drop('communityname', axis = 1)
f
```

```python
f
```

```python
from sklearn.preprocessing import StandardScaler
features = ['population', 'householdsize', 'racepctblack', 'racePctWhite']
x = f.loc[:, features].values
y = f.loc[:, ['fold']].values
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf
```

```python
finalDf = pd.concat([principalDf, f[['fold']]], axis = 1)
```

```python
finalDf
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
folds = [1, 2, 3]
colors = ['r', 'g', 'b']
for fold, color in zip(folds,colors):
    indicesToKeep = finalDf['fold'] == fold
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(folds)
ax.grid()
```

```python
import pandas as pd
import numpy as np
f2 = pd.read_excel('london-borough-profilesV2.xlsx')
f2
```

```python
f2 = f2.apply(pd.to_numeric, errors='coerce')
f2 = f2.dropna(axis=1, how='all')
f2
```

```python
f2 = f2.fillna(f2.mean())
f2
```

```python
f2 = f2 - f2.mean()
f2
```

```python
from sklearn.metrics import euclidean_distances
distMatrix = euclidean_distances(f2, f2)
distMatrix
```

```python
from sklearn.manifold import MDS
from matplotlib import pyplot as plt

mds = MDS(random_state = 0)
scaled_f2 = mds.fit_transform(f2)
print(scaled_f2)
```

```python
from sklearn import manifold
mds = manifold.MDS(
    n_components=2,
    max_iter=3000,
    eps=1e-9,
    dissimilarity="precomputed",
    n_jobs=1,
)
pos = mds.fit(distMatrix).embedding_
print(pos)
```
